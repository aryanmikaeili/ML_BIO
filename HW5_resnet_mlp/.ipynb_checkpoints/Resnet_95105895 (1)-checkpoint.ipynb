Data preperation


```python
import cv2
import imutils
def crop_imgs(set_name, add_pixels_value=0):
    """
    Finds the extreme points on the image and crops the rectangular out of them
    """
    set_new = []
    for img in set_name:
        gray = img
        if len(img.shape) == 3:
          gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        gray = cv2.GaussianBlur(gray, (5, 5), 0)
        # threshold the image, then perform a series of erosions +
        # dilations to remove any small regions of noise
        thresh = cv2.threshold(gray, 45, 255, cv2.THRESH_BINARY)[1]
        thresh = cv2.erode(thresh, None, iterations=2)
        thresh = cv2.dilate(thresh, None, iterations=2)

        # find contours in thresholded image, then grab the largest one
        cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = imutils.grab_contours(cnts)
        c = max(cnts, key=cv2.contourArea)

        # find the extreme points
        extLeft = tuple(c[c[:, :, 0].argmin()][0])
        extRight = tuple(c[c[:, :, 0].argmax()][0])
        extTop = tuple(c[c[:, :, 1].argmin()][0])
        extBot = tuple(c[c[:, :, 1].argmax()][0])

        ADD_PIXELS = add_pixels_value
        new_img = img[extTop[1]-ADD_PIXELS:extBot[1]+ADD_PIXELS, extLeft[0]-ADD_PIXELS:extRight[0]+ADD_PIXELS].copy()
        set_new.append(new_img)

    return np.array(set_new)
```

```python
croped_X = crop_imgs(X)
plt.imshow(croped_X[0], cmap = "gray")
```

```python
from torchvision import transforms
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(croped_X, Y, test_size = 0.3)
test_X, val_X, test_y, val_y = train_test_split(test_X, test_y, test_size = 0.5)

def augment(X, y, aug_cof = 10):
  transform = transforms.Compose([
    transforms.RandomRotation(5),
    transforms.RandomVerticalFlip(),
    transforms.Resize((224, 224)),
    transforms.ToTensor()

  ])
  
  new_images = []
  new_labels = []
  for i in range(aug_cof):
    for j, img in enumerate(X):
      image = Image.fromarray(img)
      image = transform(image)
      new_images.append(image.numpy())
      new_labels.append(y[j])
      print("generating image " , i * X.shape[0] + j)

  return np.array(new_images), np.array(new_labels)

train_X , train_y = augment(train_X, train_y, 60)
test_X , test_y = augment(test_X, test_y, 60)
val_X , val_y = augment(val_X, val_y, 60)
    
    

    
```

# Model classes

```python
class Block(nn.Module):
  def __init__(self, in_channel, out_channel, kernel_size, stride = 1, padding = 0, batchnorm = False, dout = False):
      super(Block, self).__init__()
      self.in_channel = in_channel
      self.out_channel = out_channel
      self.kernel_size = kernel_size
      self.stride = stride
      self.padding = padding

      self.batchnorm = batchnorm
      self.dout = dout
    
      self.conv1 = nn.Conv2d(in_channel, out_channel, kernel_size=kernel_size, stride= stride, padding = padding)
      self.conv2 = nn.Conv2d(in_channels=out_channel, out_channels=out_channel, kernel_size=kernel_size, stride = 1, padding=padding)

      self.relu = nn.ReLU(inplace = True)

      self.dropout1 = nn.Dropout(p=0.2)
      self.dropout2 = nn.Dropout(p=0.4)

      self.bn = nn.BatchNorm2d(out_channel)

      self.downsample = nn.Conv2d(in_channel, out_channel, kernel_size = 1, stride = stride, bias=False)
  def forward(self, x):
    residual = x

    o = self.conv1(x)
    if self.batchnorm:
      o = self.bn(o)

    o = self.relu(o)
    if self.dout:
      o = self.dropout2(o)
    o = self.conv2(o)
    if self.batchnorm:
      o = self.bn(o)
    
    residual = self.downsample(x)
    output = o + residual
    output = self.relu(output)
    if self.dout:
      output = self.dropout1(output)


    return output
```

```python
import math
class Resnet(nn.Module):
    def __init__(self, in_channel, batchnorm, dout):
      super(Resnet, self).__init__()
      self.in_channel = in_channel
      self.batchnorm = batchnorm
      self.dout = dout
    

      self.conv1 = nn.Conv2d(in_channel, 32, 7, 1, 3)
      self.relu = nn.ReLU(inplace = True)
      self.softmax = nn.Softmax(1)
      self.maxpool = nn.MaxPool2d(2)
      self.bn2d = nn.BatchNorm2d(32)

      self.bn1d = nn.BatchNorm1d(1024)

      self.dropout = nn.Dropout(p=0.3)

      self.block1 = Block(32, 32, 7, 2, 3, batchnorm, dout)
      self.block2 = Block(32, 64, 5, 2, 2,  batchnorm, dout)
      self.block3 = Block(64, 128, 3, 2, 1, batchnorm, dout)

      self.fc1 = nn.Linear(128 * 7 * 7, 512)
      self.fc2 = nn.Linear(512, 2)
      ###this initialization part is from resnet implementation in pytorh
      for m in self.modules():
        if isinstance(m, nn.Conv2d):
          n = m.kernel_size[0] * m.kernel_size[1] * m.out_channels
          m.weight.data.normal_(0, math.sqrt(2. / n))
 

    def forward(self, x):
      
      x = self.conv1(x)
      if self.batchnorm:
        x = self.bn2d(x)
      x = self.relu(x)
      x = self.maxpool(x)

      x = self.block1(x)
      x = self.block2(x)
      x = self.block3(x)
      x = self.maxpool(x)
      x = torch.flatten(x, 1)

      x = self.fc1(x)
      x = self.relu(x)
      x = self.fc2(x)
      #x = torch.nn.functional.softmax(x, 1)


      

      return x
    

```

# Dataset and batch generation


```python
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):
    def __init__(self,X, y):
        super(CustomDataset, self).__init__()        
        self.X = X
        self.y = y
        
        
    def __len__(self):         
        return self.X.shape[0]
    
    def __getitem__(self, idx):

        if torch.is_tensor(idx):
            idx = idx.tolist()
        return  self.X[idx], self.y[idx]

```

if you first augmented the data and then splitted it, use this block


```python
trainset = CustomDataset(X_train, y_train)

trainloader = DataLoader(trainset, batch_size=64,
                                         shuffle=False)

valset = CustomDataset(X_val, y_val)

valloader = DataLoader(trainset, batch_size=64,
                                         shuffle=False)

testset = CustomDataset(X_test, y_test)
testloader = DataLoader(testset, batch_size=64, shuffle = False)
```

if you first splitted the data and then augmented it, use this block


```python
trainset = CustomDataset(train_X, train_y)

trainloader = DataLoader(trainset, batch_size=64,
                                         shuffle=False)

valset = CustomDataset(val_X, val_y)

valloader = DataLoader(trainset, batch_size=64,
                                         shuffle=False)

testset = CustomDataset(test_X, test_y)
testloader = DataLoader(testset, batch_size=64, shuffle = False)
```

this function adjusts the learning rate. for half of the epoch I used learning rate 1e-3. for the rest I used 1e-5



```python
def adjust_learning_rate(optimizer, epoch, num_epochs):
  lr = 0
  if epoch < num_epochs // 4:
    lr = 1e-3
  elif epoch > num_epochs // 4 and epoch < num_epochs // 2:
    lr = 1e-4
  else:
    lr = 1e-5
  for param_group in optimizer.param_groups:
      param_group['lr'] = lr

```

the training function

```python
def train(net, num_epochs):

  criterion = nn.CrossEntropyLoss()
  optimizer = optim.Adam(net.parameters(), lr=1e-3)


  hist = {'training_loss': [], 'val_loss':[]}

  running_loss = 0

  for epoch in range(num_epochs):
      adjust_learning_rate(optimizer, epoch, num_epochs)
      net.train()
      training_loss = 0
      for i, data in enumerate(trainloader):
        images, labels = data[0].cuda(), data[1].cuda()
        out = net(images)
        loss = criterion(out, labels.long())
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        running_loss += loss.item()

        if i % 10 == 9:  
          print('[%d %5d] loss: %.5f' %
                (epoch, i + 1, loss.item())
          running_loss = 0.0

        training_loss += loss.item() * images.shape[0]
      net.eval()
      val_loss = 0
      with torch.no_grad():
        for i, data in enumerate(valloader):
          images, labels = data[0].cuda(), data[1].cuda()

          out = net(images)
          loss = criterion(out, labels.long())
          val_loss += loss.item() *images.shape[0]
      print( '[epoch : %4d] training loss: %.5f, validation loss: %.5f' % (epoch + 1, training_loss / len(trainset), val_loss / len(valset)))
      hist['training_loss'].append(training_loss / len(trainset))
      hist['val_loss'].append(val_loss / len(valset))
  return hist


    

```

# the model that doesn't have dropout and batch normalization




```python
net = Resnet(1, False, False)
device = torch.device("cuda:0")
net.to(device)
loss_log = train(net, 5)
```

```python
plt.plot(loss_log["training_loss"], label = "training loss")
plt.plot(loss_log["val_loss"], label = "validation loss")
plt.legend()
plt.show()
```

# Accuracy of the model without dropout and batch norm

```python
num_true = 0

net.eval()
with torch.no_grad():
  for data in testloader:
        images,labels = data[0].cuda(), data[1].cuda()
        outputs = net(images)
        preds = torch.argmax(outputs, 1)
        num_true += np.where((preds == labels).cpu().numpy() == True)[0].shape[0]

print("accuracy: ",num_true / len(testset))
```
